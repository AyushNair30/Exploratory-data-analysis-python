```python
import pandas as pd
import seaborn as sns

df=pd.read_csv("TWO_CENTURIES_OF_UM_RACES.csv")
```

#Checking the imported data

```python
df.shape
```

```python
df.dtypes
```

#Clean up data


#Only want USA races ,50km or 50 mi ,2020

```python
#Step1: Show 50km or 50 mi
#50km
#50mi

df[df['Event distance/length']=='50mi']
```

```python
#combine  50km and 50mi with isin

df[df['Event distance/length'].isin(['50km', '50mi'])]
```

```python
df[(df['Event distance/length'].isin(['50km', '50mi']))   &  (df['Year of event']==2020)]
```

```python
df[df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0)=='USA'].head()
```

```python
#combine all filters

df[(df['Event distance/length'].isin(['50km', '50mi']))   &  (df['Year of event']==2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0)=='USA')]
```

```python
df2=df[(df['Event distance/length'].isin(['50km', '50mi']))   &  (df['Year of event']==2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0)=='USA')]
```

```python
df2.shape
```

```python
#removing USA from event name

df2['Event name']=df2['Event name'].str.split('(').str.get(0)
```

```python
#clean up athlete age

df2['Athlete age']=2020-df2['Athlete year of birth']
```

```python
#remove 'h' from athlete performance

df2["Athlete performance"]=df2["Athlete performance"].str.rstrip('h')
```

```python
df2.head()
```

```python
#drop Athlete Club,Athlete Year Of Birth,Athlete Country,Athlete age category 

df2.drop(['Athlete club','Athlete year of birth','Athlete country','Athlete age category'],axis=1,inplace=True)
```

```python
df2.head()
```

```python
#clean up null values

df2.isna().sum()
```

```python
df2[df2['Athlete age'].isna()==1]
```

```python
df2=df2.dropna()
```

```python
#check for duplicate values

df2[df2.duplicated()==True]
```

```python
#reset index

df2.reset_index(inplace=True,drop=True)
```

```python
#fix types

df2.dtypes
```

```python
df2['Athlete age']=df2['Athlete age'].astype(int)
```

```python
df2['Athlete average speed']=df2['Athlete average speed'].astype(float)
```

```python
#rename columns

df2=df2.rename(columns={'Year of event':'year','Event name':'race_name','Event dates':'race_day','Event distance/length':'race_length','Athlete performance':'athlete_performance','Athlete average speed':'athlete_average_speed','Athlete ID':'athlete_id','Athlete gender':'athlete_gender'})
```

```python
df2.head()
```

#Additional Questions

```python
#difference in speed for the 50k,50mi male to female

df2.groupby(['race_length', 'athlete'])['athlete_average_speed'].mean()
```

```python
#Split by seasons
#spring 3-5
#summer 6-8
#fall 9-11
#winter 12-2

df2['race_month']=df2['race_day'].str.split('.').str.get(1).astype(int)
```

```python
df2.head()
```

```python
df2['race_season']=df2['race_month'].apply(lambda x:'spring' if x in [3,4,5] else 'summer' if x in [6,7,8] else 'fall' if x in [9,10,11] else 'winter')
```

```python
df2
```

```python
df2.groupby('race_season')['athlete_average_speed'].agg(['mean','count']).sort_values('mean',ascending=False)
```
